```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta


def calculate_engagement_rate(total_engagements, total_followers):
    return (total_engagements / total_followers) * 100


data = {
    "date": pd.date_range(end=datetime.today(), periods=30),
    "followers": [1000 + i * 5 for i in range(30)],
    "likes": [100 + i * 2 for i in range(30)],
    "comments": [20 + i for i in range(30)],
    "shares": [10 + i for i in range(30)],
    "post_engagements": [130 + i * 3 for i in range(30)],
}


df = pd.DataFrame(data)


df["engagement_rate"] = calculate_engagement_rate(df["post_engagements"], df["followers"])


engagement_rate_last_month = df["engagement_rate"].mean()
follower_growth = df["followers"].iloc[-1] - df["followers"].iloc[0]
top_posts = df.nlargest(5, "post_engagements")


plt.figure(figsize=(10, 6))
plt.plot(df["date"], df["engagement_rate"], marker="o")
plt.title("Engagement Rate Over Last Month")
plt.xlabel("Date")
plt.ylabel("Engagement Rate (%)")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


print("Social Media Performance Insights:")
print(f"- Average Engagement Rate Over Last Month: {engagement_rate_last_month:.2f}%")
print(f"- Growth in Followers: {follower_growth}")
print("\nTop-Performing Posts (Last Month):")
print(top_posts[["date", "post_engagements"]])

print("\nRecommendations:")
print("1. Focus on posting content similar to top-performing posts to maintain engagement.")
print("2. Optimize posting times to match when followers are most active.")
print("3. Engage directly with followers through comments and stories.")
print("4. Use targeted ads to attract followers from underrepresented demographics.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Placeholder data for three competitors
data = {
    "competitor": [
        "Competitor A", "Competitor A", "Competitor A",
        "Competitor B", "Competitor B", "Competitor B",
        "Competitor C", "Competitor C", "Competitor C"
    ],
    "post_date": [
        "2025-01-01", "2025-01-10", "2025-01-15",
        "2025-01-02", "2025-01-08", "2025-01-14",
        "2025-01-03", "2025-01-12", "2025-01-18"
    ],
    "engagement_rate": [5.2, 6.8, 7.1, 4.5, 5.0, 5.6, 6.0, 6.5, 7.0],
    "hashtags": [["#style", "#trend"], ["#fashion", "#ootd"], ["#lookbook"]],
    "visual_consistency": ["high", "high", "medium", "medium", "low", "low", "high", "high", "medium"],
}

# Create a DataFrame
df = pd.DataFrame(data)

# Analyzing Frequency of Posts
post_frequency = df["competitor"].value_counts()

# Analyzing Average Engagement Rates
average_engagement = df.groupby("competitor")["engagement_rate"].mean().reset_index()
average_engagement = average_engagement.sort_values(by="engagement_rate", ascending=False)

# Analyzing Hashtag Usage
df["hashtag_count"] = df["hashtags"].apply(len)
average_hashtags = df.groupby("competitor")["hashtag_count"].mean().reset_index()

# Analyzing Visual Consistency
visual_consistency = df.groupby("competitor")["visual_consistency"].value_counts(normalize=True).unstack().fillna(0)

# Visualization: Engagement Rates
plt.figure(figsize=(8, 5))
plt.bar(average_engagement["competitor"], average_engagement["engagement_rate"], color="lightblue")
plt.title("Average Engagement Rates by Competitor")
plt.xlabel("Competitor")
plt.ylabel("Engagement Rate (%)")
plt.show()

# Visualization: Visual Consistency
visual_consistency.plot(kind="bar", stacked=True, figsize=(8, 5), colormap="coolwarm")
plt.title("Visual Consistency by Competitor")
plt.xlabel("Competitor")
plt.ylabel("Proportion")
plt.legend(title="Consistency Level")
plt.tight_layout()
plt.show()

# Insights
print("Competitor Analysis Insights:\n")
print("1. Frequency of Posts (Last 2 Weeks):")
print(post_frequency)

print("\n2. Average Engagement Rates:")
print(average_engagement)

print("\n3. Average Number of Hashtags Used per Post:")
print(average_hashtags)

print("\n4. Visual Consistency Analysis:")
print(visual_consistency)

# Recommendations
print("\nRecommendations for Differentiation:")
print("1. Post more consistently than competitors with visually cohesive content.")
print("2. Use hashtags strategically by mixing high-performing niche and trending hashtags.")
print("3. Leverage unique visual styles and interactive content to stand out.")
print("4. Focus on high-engagement formats like videos, live streams, and carousel posts.")
```
